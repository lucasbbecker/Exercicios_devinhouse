## Exercício 01

```python
import pandas as pd

clientes = pd.read_csv("clientes.csv")
pedidos = pd.read_csv("pedidos.csv")
produtos = pd.read_csv("produtos.csv")

print("CLIENTES")
display(clientes.head())

print("PEDIDOS")
display(pedidos.head())

print("PRODUTOS")
display(produtos.head())


print("CLIENTES - info()")
clientes.info()
print("\nCLIENTES - nulos")
print(clientes.isnull().sum())

print("\nPEDIDOS - info()")
pedidos.info()
print("\nPEDIDOS - nulos")
print(pedidos.isnull().sum())

print("\nPRODUTOS - info()")
produtos.info()
print("\nPRODUTOS - nulos")
print(produtos.isnull().sum())

print("CLIENTES - describe()")
display(clientes.describe(include="all"))

print("\nPEDIDOS - describe()")
display(pedidos.describe(include="all"))

print("\nPRODUTOS - describe()")
display(produtos.describe(include="all"))

```

## Exercício 04

```python
import numpy as np

valores_pedidos = pedidos["valor_total"].to_numpy()
valores_pedidos

# Calcular média, soma e desvio padrão
media_valores = np.mean(valores_pedidos)
soma_valores = np.sum(valores_pedidos)
desvio_padrao_valores = np.std(valores_pedidos)

print("Média dos pedidos:", media_valores)
print("Soma dos pedidos:", soma_valores)
print("Desvio padrão dos pedidos:", desvio_padrao_valores)

# Aplicar desconto de 10% com broadcasting
valores_com_desconto = valores_pedidos * 0.9
valores_com_desconto

# Verificar tipos e dimensões com .dtype e .shape
print("dtype do array:", valores_pedidos.dtype)
print("shape do array:", valores_pedidos.shape)

```

## Exercício 05

```python
# Total gasto por cliente
total_por_cliente = pedidos.groupby("id_cliente")["valor_total"].sum().reset_index()

# Juntar com a tabela de clientes
clientes_total = clientes.merge(total_por_cliente, on="id_cliente", how="left")

# Substituir NaN por 0 caso algum cliente nunca tenha comprado
clientes_total["valor_total"] = clientes_total["valor_total"].fillna(0)

# Encontrar o cliente que mais gastou
cliente_top = clientes_total.loc[clientes_total["valor_total"].idxmax()]
cliente_top

ticket_medio_categoria = produtos.groupby("categoria")["preco"].mean().reset_index()

ticket_medio_categoria

clientes_por_cidade = clientes.groupby("cidade")["id_cliente"].count().reset_index()

clientes_por_cidade.rename(columns={"id_cliente": "quantidade_clientes"}, inplace=True)

cidade_top = clientes_por_cidade.loc[clientes_por_cidade["quantidade_clientes"].idxmax()]

clientes_por_cidade, cidade_top

import matplotlib.pyplot as plt

# Gráfico de barras - clientes por cidade
plt.figure(figsize=(10,5))
plt.bar(clientes_por_cidade["cidade"], clientes_por_cidade["quantidade_clientes"])
plt.title("Quantidade de clientes por cidade")
plt.xlabel("Cidade")
plt.ylabel("Clientes")
plt.xticks(rotation=45)
plt.show()
```
